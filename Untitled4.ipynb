```python
import pandas as pd
import numpy as np
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
final_dataset= pd.read_csv("final_dataset.csv")
```

```python
final_dataset
```

```python
final_dataset.isnull().sum()
```

```python
target=final_dataset["vote_average"]
```

```python
feature= final_dataset.drop(["title","actor_1","actor_2","actor_3", "actor_4","actor_5", "actor_6","actor_7","actor_8", "actor_9", "actor_10", "actor_11"],axis=1)
```

```python
vectorizer= TfidfVectorizer(ngram_range=(1,2), lowercase=False, stop_words=None, tokenizer=None)
```

```python
import string
import codecs
```

```python
translate_text= dict((ord(char), None)for char in string.punctuation)
```

```python
genre_feature=[]
for i,j in feature.iterrows():
    j= j["genre"]
    j=j.lower()
    genre_tokens=" ".join(word_tokenize(j))
    genre_tokens= genre_tokens.replace("|"," ")
    genre_feature.append(genre_tokens)
    
```

```python
genre_feature
```

```python
feature["Oerview"]=feature["Oerview"].fillna("unknown")
feature["new_key"]=feature["new_key"].fillna("unknown")
```

```python
new_feature= []
new_feature2= []
for i, j in feature.iterrows():
    key_s= j["new_key"]
    text=j["Oerview"]
    text= text.lower()
    text= re.sub(r'\d+'," ", text)
    tokens= " ".join(word_tokenize(text))
    new_feature.append(tokens)
    key_s= key_s.lower()
    key_s= re.sub(r'\d+'," ", key_s)
    key_tokens= " ".join(word_tokenize(key_s))
    new_feature2.append(key_tokens)
    
```

```python
from sklearn.model_selection import train_test_split
```

```python
first= pd.DataFrame(new_feature, columns=["overview"])
second= pd.DataFrame(new_feature2, columns=["keywords"])
third= pd.DataFrame(genre_feature, columns=["genre"])
```

```python
fourth=pd.concat([first, second], axis=1)

fifth= pd.concat([fourth, third], axis=1)
```

```python
fifth.drop_duplicates(keep="first", inplace=True)
```

```python
fifth
```

```python
x_train_full=fifth["genre"]+fifth["keywords"]
```

```python
x_train_full
```

```python
x_train, x_test, y_train, y_test= train_test_split(x_train_full[:9000], target[:9000], test_size=0.5, random_state=42)
```

```python
x_train_vectorized= vectorizer.fit_transform(x_train).toarray()
```

```python
x_test_vectorize= vectorizer.transform(x_test).toarray()
```

```python
from sklearn.svm import SVR
```

```python
svr= SVR()
```

```python
%%time
svr.fit(x_train_vectorized, y_train)
```

```python
pred=svr.predict(x_test_vectorize[:1500] )
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
score=mean_squared_error(y_test[:1500], pred)
```
