## it's time to extract the genres


```python
import ast
```

```python
metadata["new_genre"]= metadata["genres"].apply(lambda x:ast.literal_eval(x))
```

```python
def genre_extract(names):
    genre=[]
    st="|"
    for i in names:
        if i.get('name'=='Science Fiction'):
            genre.append("Sci-Fi")
        else:
            genre.append(i.get("name"))
            
    if genre==[]:
        return np.NaN
    else:
        return st.join(genre)
```

```python
final_metadata= pd.DataFrame(metadata.id)
```

```python
final_metadata["genre"]= metadata["new_genre"].map(lambda x: genre_extract(x))
```

```python
metadata.columns
```

```python
final_metadata["title"]= metadata["title"]
final_metadata["runtime"]= metadata["runtime"]
final_metadata["release_date"]= metadata["release_date"]

final_metadata["vote_average"]= metadata["vote_average"]





```

```python
final_metadata
```

```python
import datetime as dt
```

```python
final_metadata["release_date"]=pd.to_datetime(final_metadata["release_date"]).dt.date
```

```python
final_metadata["runtime"]= pd.to_datetime(final_metadata["runtime"], unit='m').dt.time
```
